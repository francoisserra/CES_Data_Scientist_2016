# I - Prétraitement et visualiation sur *digits*


Nous passons maintenant à un jeu de données classique en machine learning dans le cadre de la classification de chiffres manuscrits.

Après echantillonnage, on observe de l'aliasing au niveau des hautes frequences

## Description du jeu de données:
On charge le jeu de données *digits* disponoible dans le package scikit-learn (nom d'import sklearn). Ce jeu de données contient des images de chiffres numérisés. On va s'en servir au cours de cette séance pour étudier les principaux enjeu en classification (supervisée).  

```python
# Chargement des données disponible dans le package scikit-learn
from sklearn.datasets import load_digits

digits = load_digits()
X, y = digits.data, digits.target

print("Nombre de pixels :      {}".format(X.shape[1]))
print("Nombre d'observations : {}".format(X.shape[0]))
print("Nombre de classes :     {}".format(len(np.unique(y))))

# Choix d'une observation quelconques de la base
idx_to_test = 15

print("Affichage d'une ligne de la matrice / image:")
print(X[idx_to_test, :])
print("Affichage de la classe / chiffre associé:")
print(y[idx_to_test])
```


## Statistiques élementaires :
Pour mieux comprendre la base de données on va s'intéresser à quelques statistiques. 
On commence par calculer les moyennes et variances par classes pour chacun des chiffres. La moyenne par classe se visualise comme une image qui est une représentantion moyenne pour chaque chiffre de zéro à neuf. Idem pour la variance, ce qui permet alors de voir les parties avec les plus grandes variations entre les membres d'une même classe.


```python
# Récupérer les modalités possible prises (Il y en a bien 10!) 
classes_list = np.unique(y).astype(int)
print "Liste des classes en présence: ", classes_list
```

**Questions**:

- Calculer un représentant moyen du chiffre 0? Utiliser la fonction `np.mean`
- Avec une boucle `for` calculer et visualiser le représentant moyen pour chaque chiffre
- Calculer la variance du chiffre 0? Utiliser la fonction `np.var`
- Avec une boucle `for` calculer et visualiser la variance pour chaque chiffre

```python
import numpy as np
centroid_class0 = np.mean(X[y.ravel()==0,:],axis=0)
plt.title(u'centroid class 0')
plt.imshow(centroid_class0.reshape((8,8)),cmap='gray', aspect='equal', interpolation='nearest')
plt.axis('off')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=10, figsize=(20,4))
for i,ax in enumerate(axes):
    centroid_class = np.mean(X[y.ravel()==i,:],axis=0)
    ax.set_title("centroid class "+str(i))
    ax.imshow(centroid_class.reshape((8,8)),cmap='gray', aspect='equal', interpolation='nearest')
plt.axis('off')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=10, figsize=(20,4))
for i,ax in enumerate(axes):
    variance_class = np.var(X[y.ravel()==i,:],axis=0)
    ax.set_title("variance class "+str(i))
    ax.imshow(variance_class.reshape((8,8)),cmap='gray', aspect='equal', interpolation='nearest')
plt.axis('off')
plt.show()
```

# II - Mon premier classifieur

## Validation : découpage Apprentissage / Validation 

On procède de manière classique en réservant 80% des données pour la partie apprentissage, et 20% pour l'évaluation des classifieurs que l'on a construit sur la première partie.
En effet il n'est pas raisonnable de tester la perfromance sur 100% des données. Cela donnera lieux à du sur-apprentissage (en: overfitting). La généralisation des méthodes apprises serait alors très mauvaise

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8,
                                                    random_state=42)
print("Nb d'échantillons d'apprentissage :  {}".format(X_train.shape[0]))
print("Nb d'échantillons de validation :    {}".format(X_test.shape[0]))
```

```python
plt.hist(y_train, bins=10)
plt.hist(y_test, bins=10)
```

## Classification par plus proche centroide

**Questions**:

- Pour chaque chiffre calculer sur l'ensemble de train le chiffre moyen. On notera la variable contenant les moyennes `centroids_train`
- Pour chaque chiffre de l'ensemble de test, calculer le centroide le plus proche. Vous évaluerez si le chiffre ainsi obtenu correspond au vrai chiffre et en déduirez une estimation du taux de bonnes prédictions.

```python
from itertools import izip

centroids_train = np.zeros((10,X.shape[1]))
for i in np.arange(10):
    centroids_train[i,:] = np.mean(X_train[y_train.ravel()==i,:],axis=0)

y_prediction = np.zeros((np.shape(y_test)))
for i,(xt,yt) in enumerate(izip(X_test,y_test)):
    distances = np.linalg.norm(centroids_train-xt,axis=1)
    y_prediction[i] = np.argmin(distances) 
        
confusion_matrix = np.zeros((10,10))
number_of_elements_per_class = np.zeros((10))
for j in np.arange(10):
    number_of_elements_per_class[j] = np.sum(y_test==j)
for i in np.arange(10):
    for j in np.arange(10):
        confusion_matrix[i,j] = np.sum(y_prediction[y_test==i]==j)*100/number_of_elements_per_class[i]
for i in np.arange(10):
    print u'recognition rate for class '+str(i)+": ", confusion_matrix[i,i],"%"


plt.title(u'confusion matrix (custom)')
plt.imshow(confusion_matrix,cmap='spectral', aspect='equal', interpolation='nearest')
plt.colorbar()
plt.axis('off')
plt.show()
```

```python
from sklearn.neighbors.nearest_centroid import NearestCentroid

clf = NearestCentroid()
clf.fit(X_train,y_train)
print "error between custom centroids and sklearn centroids : ",np.mean((clf.centroids_-centroids_train)**2)

y_prediction = clf.predict(X_test) 

for i in np.arange(10):
    for j in np.arange(10):
        confusion_matrix[i,j] = np.sum(y_prediction[y_test==i]==j)*100/number_of_elements_per_class[i]
for i in np.arange(10):
    print u'recognition rate for class '+str(i)+": ", confusion_matrix[i,i],"%"


plt.title(u'confusion matrix (scikit)')
plt.imshow(confusion_matrix,cmap='spectral', aspect='equal', interpolation='nearest')
plt.colorbar()
plt.axis('off')
plt.show()
```
